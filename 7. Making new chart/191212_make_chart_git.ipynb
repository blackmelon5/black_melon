# 새로운 차트를 위한 score를 정하기 위해 min-max scale을 이용해 스케일링하기

```python
melon = pd.read_csv('191212_melon_chart_all')
```

```python
like_zero = melon[melon['like'] == 0].index.tolist()
```

```python
melon['like'][78]
```

```python
# like 0으로 들어간 거 
for i in like_zero:
    melon['like'][i] = melon['like'].mean()
```

```python
melon['album_like/like'] = melon['album_like']/melon['like']
```

```python
melon['fan_per'] = minmax_scale(np.array(melon['fan_max']), axis=0, copy=True)
```

```python
melon
```

```python
melon['album_reply_per'] = minmax_scale(np.array(melon['album_reply']), axis=0, copy=True)
melon['reply_per'] = minmax_scale(np.array(melon['reply']), axis=0, copy=True)
melon['like_per'] = minmax_scale(np.array(melon['like']), axis=0, copy=True)
```

```python
melon['album_score_per'] = minmax_scale(np.array(melon['album_score']), axis=0, copy=True)
melon['album_score_count_per'] = minmax_scale(np.array(melon['album_score_count']), axis=0, copy=True)
melon['user_per_per'] = minmax_scale(np.array(melon['user_per']), axis=0, copy=True)
```

```python
melon['album_like_per'] = minmax_scale(np.array(melon['album_like']), axis=0, copy=True)
```

```python
melon['album_like_like_per'] = minmax_scale(np.array(melon['album_like/like']), axis=0, copy=True)
```

```python
melon
```

# feature importance 이용해서 가중치 두기

```python
melon['score'] = melon['user_per_per'] * 0.7 + (melon['fan_per'] * 0.06908332 + melon['like_per']*0.13545797 + melon['album_score_per'] * 0.54701395 + melon['album_like_like_per']*0.24844475) 
```

```python
melon_new_chart = melon.sort_values('score', ascending = False)
```
