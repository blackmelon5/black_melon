# 가설 : 어뷰징 가수는 팬 맺은 수가 적을 것이다

```python
fig = px.scatter(melon_title[melon_title['user_per']>5], x = 'user_per', y = 'fan_max',
                 color = 'abusing', size = 'album_score_count', hover_data = ['artist', 'title'],color_continuous_scale = 'Viridis')
fig.update_layout(plot_bgcolor='rgb(240, 240, 240)')
fig.show()
```

# 가설 : 어뷰징 가수는 음원 좋아요에 비해 음원 좋아요의 비율이 적을 것이다

```python
px.scatter(melon_title[melon_title['user_per']>5], x = 'like', y = 'album_like',color = 'abusing', size = 'user_per', hover_data = ['artist', 'title'])
```

- 노 어뷰징 가수는 linear한 편이지만 어뷰징 가수는 거의 일자형

```python
fig =  px.scatter(melon_title[(melon_title['user_per']>5)], 
                 x = 'like', y = 'album_like', size = 'user_per', color = 'abusing',facet_col='abusing',
                  hover_data = ['artist', 'title'], trendline = 'ols')
fig.update_traces(
    line=dict(dash="dot", width=4),
    selector=dict(type="scatter", mode="lines"))

fig.update_layout(plot_bgcolor='rgb(240, 240, 240)')
fig.show()
```

- 어뷰징 가수의 기울기는 아주 낮은 반면에
- 노어뷰징 가수들의 linear하게 올라가는 걸 볼 수 있다
- 음원 좋아요가 높은 어뷰징 가수(차트에 오래 남아 데이터가 희석된 가수)들도 앨범 좋아요 수는 낮은 것을 볼 수 있다.

```python
melon_title['like_albumlike_ratio'] = melon_title['album_like']/melon_title['like']
```

```python
px.scatter(melon_title[melon_title['user_per']>5], x = 'like', y = 'like_albumlike_ratio',color = 'abusing',size = 'album_like', hover_data = ['artist', 'title'])
```

- x축을 음원 좋아요로 두고 y축을 음원대비 앨범 좋아요수 비율로 두었다. size 는 앨범 좋아요로.
- 어뷰징 가수들은 거의 0.1 아래에 위치해있다.
- 어뷰징을 할 때 음원 하트수는 신경을 써도 앨범 좋아요 수는 신경을 못 쓴다는 증거이며
- 인기가 많은 노래일 경우 앨범 하트수도 크다는 것이 일반적이나 어뷰징 가수는 일반적이지 않으므로 낮은 비율일 거라는 가설의 증거가 될 수 있다.

```python
melon_title = melon_title.reset_index().drop(columns = 'index')
```

```python
# 평균 순위 구하기, 역대차트순위합/차트에든횟수
melon_title['chart_trend/count'] = [sum(melon_title['chart_trend'][i])/melon_title['count'][i] for i in range(len(melon_title))]
```

```python
px.scatter(melon_title[(melon_title['fan_max'] < 20000) & (melon_title['album_release_date']> '2019-01-01')], x='fan_max', y='chart_trend/count', color = 'abusing', hover_data=['abusing', 'artist', 'title'])
```

- y축은 평균 순위, x축은 팬 수
- 같은 팬수여도 평균순위가 굉장히 높음(0에 가까울수록 높은 순위)

```python
px.scatter( melon_title[(melon_title['abusing'] ==1) & (melon_title['album_release_date']> '2019-01-01')], x='fan_max', y='chart_trend/count', hover_data=['abusing', 'artist', 'title'])
```

```python
px.scatter( melon_title[((melon_title['fan_max'] > 100000) & (melon_title['abusing'] ==0) & (melon_title['album_release_date']> '2019-01-01')) | ((melon_title['fan_max'] < 20000) & (melon_title['abusing'] ==1) & (melon_title['album_release_date']> '2019-01-01'))], x='user_per', y='chart_trend/count', color = 'abusing', hover_data=['abusing', 'artist', 'title'])
```

- 2019년에 나온 팬수가 10만 이상이고 노어뷰징 가수와 어뷰징 가수(팬수가 거의 2만 아래)와 비교해보니 특징 없이 잘 섞임
